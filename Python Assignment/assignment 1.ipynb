```python
# Pattern (a)
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

```python
rows = 4
for i in range(rows):
    print(" " * (rows - i - 1) + "*" * (2 * i + 1))
```

```python
rows = 3

# Upper part
for i in range(rows):
    print(" " * (rows - i - 1) + "*" * (2 * i + 1))

# Lower part
for i in range(rows - 2, -1, -1):
    print(" " * (rows - i - 1) + "*" * (2 * i + 1))
```

```python
rows = 4
cols = 5

for i in range(rows):
    for j in range(cols):
        if i == 0 or i == rows - 1 or j == 0 or j == cols - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

```python
# Small hollow diamond pattern

# half height
n = 2

# Upper part
print(" " * (n) + "* *")

# Middle part
print(" " * (n + 1) + "*")

# Lower part
print(" " * (n) + "* *")
```
