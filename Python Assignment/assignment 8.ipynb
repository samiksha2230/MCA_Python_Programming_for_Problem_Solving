```python
import os

# Take directory path from user
directory = input("Enter directory path: ")

# Get the list of files
files = os.listdir(directory)

# Dictionary to store categorized files
categorized = {}

for file in files:
    if os.path.isfile(os.path.join(directory, file)):
        # Extract extension
        if '.' in file:
            ext = file.split('.')[-1]
            categorized.setdefault(ext, []).append(file)

# Print output one below the other
for ext, file_list in categorized.items():
    print(f"{ext}:")
    for f in file_list:
        print("  -", f)
```
