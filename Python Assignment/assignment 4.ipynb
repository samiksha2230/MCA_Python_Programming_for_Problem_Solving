```python
# Program to search for an item in a user-provided list
# Taking list input from user
items = input("Enter items separated by spaces: ").split()
print("Enterd item:", items)
```

```python
# Taking item to search
search_item = input("Enter the item to search: ")
print("Enter the item to search:",search_item)
```

```python
# Searching for the item
found = False
for i in range(len(items)):
    if items[i] == search_item:
        print(f"Item found at position {i + 1}")
        found = True
        break
if not found:
    print("Item not found.")
```
