```python
text = input("Enter a sentence: ")
# Display original text
print("Original sentence:\n", text)

# Convert to lowercase
text = text.lower()

# List of punctuation marks to remove (manually defined)
punctuations = ['.', ',', '!', '?', ';', ':', '"', "'", '-', '(', ')']

# Remove punctuation
for p in punctuations:
    text = text.replace(p, "")

# Split into words
words = text.split()

# Count frequencies
freq = {}
for w in words:
    freq[w] = freq.get(w, 0) + 1

print(freq)
```
