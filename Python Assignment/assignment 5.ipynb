```python
# Program to sort employee records by salary

# Sample list of employee dictionaries
employees = [
    {"name": "John", "salary": 45000},
    {"name": "Alex", "salary": 48000},
    {"name": "Mary", "salary":52000},
]

# Sorting by salary (ascending)
sorted_employees = sorted(employees, key=lambda x: x["salary"])

# Displaying sorted list
print("Employees sorted by salary:")
for emp in sorted_employees:
    print(f"Name: {emp['name']}, Salary: {emp['salary']}")
```
