```python
# Create a text file and write content into it
with open("sample.txt", "w") as file:
    file.write("Python makes programming fun.")
```

```python
lines = 0
words = 0
characters = 0

# Display file content
print("File Content:")
with open("sample.txt", "r") as f:
    content = f.read()
print(content)

# Open and read file
with open("sample.txt", "r") as file:
    for line in file:
        lines += 1
        characters += len(line)
        words += len(line.split())

print("Lines:", lines)
print("Words:", words)
print("Characters:", characters)
```
