```python
# Generate a random password of 12 characters
# Includes uppercase, lowercase, digits, and special characters
# Starts with a capital letter or digit

import random

# Character sets
UPPERCASE = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
lowercase = "abcdefghijklmnopqrstuvwxyz"
digits = "0123456789"
special = "%$#!"

# Step 1: Start with a capital letter or digit
first_char = random.choice(UPPERCASE + digits)

# Step 2: Generate the remaining 11 characters
all_char = UPPERCASE + lowercase + digits + special
password = first_char + ''.join(random.choice(all_char) for _ in range(11))

# Output the generated password
print("Generated Password:", password)
```
