```python
# Create a list of students, randomly select the Class Representative (CR), and shuffle the list

import random

# List of students
students = ["Aashi", "Aman", "Sagar", "Shreya"]

# Randomly select a Class Representative
cr = random.choice(students)

# Shuffle the student list
random.shuffle(students)

# Display the results
print("Class Representative (CR):", cr)
print("Shuffled Student List:", students)
```
