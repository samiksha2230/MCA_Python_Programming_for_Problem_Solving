```python
# Program to count the number of vowels in a string

# Input from user
string = input("Enter a string: ")
print("enterd string:", string)
```

```python
# Define vowels
vowels = "aeiouAEIOU"

# Initialize counter
count = 0

# Loop through each character in the string
for char in string:
    if char in vowels:
        count += 1

# Display result
print("Number of vowels in the string:", count)
```
