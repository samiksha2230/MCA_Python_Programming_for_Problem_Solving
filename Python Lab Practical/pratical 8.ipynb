```python
# Program to find the student with the highest marks (show all input)

students = {}

n = int(input("Enter number of students: "))

for i in range(n):
    name = input(f"Enter name of student {i+1}: ")

    # Validate marks
    while True:
        marks_input = input(f"Enter marks of {name}: ")
        try:
            marks = float(marks_input)
            break
        except ValueError:
            print("Invalid entry! Please enter a valid number.")

    students[name] = marks

# --- Show all inputs ---
print("\nAll students and their marks:")
for name, marks in students.items():
    print(f"{name} -> {marks}")

# --- Find highest ---
top_student = max(students, key=students.get)
top_marks = students[top_student]

print("\nStudent with the highest marks:")
print(f"{top_student} -> {top_marks}")
```
